## Question 1

Prepare Energy Indicators dataset:

```python
energy = pd.read_excel("Energy Indicators.xls", header=None) # Load Excel sheet

energy = energy.loc[18:243] # Remove unwanted rows
energy = energy.drop([0, 1], axis = 1) # Remove unwanted columns
energy.columns = ["Country", "Energy Supply", "Energy Supply per Capita", "% Renewable"] # add header

# Replace non-numeric values with NaN
for col in energy.columns[1:]:
    energy[col] = pd.to_numeric(energy[col], errors="coerce")

# Convert Energy Supply to Gigajouls
energy["Energy Supply"] *= 1e6

# Remove name parts that are in parenthesis
energy["Country"] = energy["Country"].apply(lambda name: name.split(" (")[0])

# Remove numbers from country names
energy["Country"] = energy["Country"].apply(lambda name: "".join(char for char in name if not char.isdigit()))

# Renaming certain countries
countries_rename = [
    {"old": "Republic of Korea", "new": "South Korea"},
    {"old": "United States of America", "new": "United States"},
    {"old": "United Kingdom of Great Britain and Northern Ireland", "new": "United Kingdom"},
    {"old": "China, Hong Kong Special Administrative Region", "new": "Hong Kong"}
]
for names in countries_rename:
    energy.loc[energy["Country"] == names["old"], "Country"] = names["new"]

energy
```

Prepare World Bank GDP dataset:

```python
GDP = pd.read_csv("world_bank.csv", skiprows=4) # Load CSV

# Renaming certain countries
countries_rename = [
    {"old": "Korea, Rep.", "new": "South Korea"},
    {"old": "Iran, Islamic Rep.", "new": "Iran"},
    {"old": "Hong Kong SAR, China", "new": "Hong Kong"}
]
for names in countries_rename:
    GDP.loc[GDP["Country Name"] == names["old"], "Country Name"] = names["new"]

GDP
```

Prepare Sciamo Journal dataset:

```python
ScimEn = pd.read_excel("scimagojr.xlsx")

ScimEn
```

Joining datasets:

```python
def answer_one():

    answer_header = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

    answer = pd.DataFrame(columns=answer_header[1:1])
    answer["Rank"] = ScimEn["Rank"][:15]
    answer.index = ScimEn["Country"][:15]

    for col_name in answer_header[1:]:
        if col_name in ScimEn.columns:
            answer = pd.merge(answer, ScimEn[["Country", col_name]], left_index=True, right_on="Country", how="left").set_index(answer.index).drop("Country", axis=1)

    for col_name in answer_header:
        if col_name in energy.columns:
            answer = pd.merge(answer, energy[["Country", col_name]], left_index=True, right_on="Country", how="left").set_index(answer.index).drop("Country", axis=1)

    for col_name in answer_header:
        if col_name in GDP.columns:
            answer = pd.merge(answer, GDP[["Country Name", col_name]], left_index=True, right_on="Country Name", how="left").set_index(answer.index).drop("Country Name", axis=1)

    return answer

answer_one()
```

## Question 2

```python
def answer_two():
    top15 = answer_one()
    start_year = 2006
    years = 10
    return sum(top15[str(i)] for i in range(start_year, start_year + years)) / years

answer_two()
```

## Question 3

```python
def answer_three():
    country = answer_two().nlargest(6).index[-1]
    country_row = answer_one().loc[country]
    return country_row["2015"] - country_row["2006"]

answer_three()
```

```python
def answer_six():
    top15 = answer_one()
    top15["Citable Documents per Capita"] = top15["Citable documents"] / top15["Energy Supply"] * top15["Energy Supply per Capita"]
    return top15["Citable Documents per Capita"].corr(top15["Energy Supply per Capita"])

answer_six()
```

## Question 7

```python
continentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

def answer_seven():

    top15 = answer_one()
    top15["Population"] = top15["Energy Supply"] / top15["Energy Supply per Capita"]

    cont_names = []
    for key in continentDict.keys():
        cont = continentDict[key]
        if cont not in cont_names: cont_names.append(cont)

    top15["Continent"] = top15.index.to_series().apply(lambda c: continentDict[c])

    continents = pd.DataFrame(columns=['size', 'sum', 'mean', 'std'])
    for continent in cont_names:
        cont_rows = top15[top15["Continent"] == continent]
        size = len(cont_rows)
        popul_sum = sum([top15.loc[country]["Population"] for country in cont_rows.index])
        mean = popul_sum / size
        std_dev = (sum([(top15.loc[country]["Population"] - mean)**2 for country in cont_rows.index])/size)**0.5

        continents.loc[continent] = [size, popul_sum, mean, std_dev]
    
    return continents

answer_seven()
```
