First, we create a table of all possible combinations (a, b, their sum and product), with the following conditions:
 - 1 < a, b < 100
 - a + b <= 100
 - a != b
 - no repetitions - ensured by setting a < b

```python
combinations = []

for a in range(2, 100):
    for b in range(2, 100):
        if a < b and a + b <= 100:
            combinations.append(
                {
                    "a": a,
                    "b": b,
                    "sum": a + b,
                    "prod": a * b
                }
        )

combinations
```

Now, we determine all of the unique sums that are possible from this set of combinations

```python
unique_sums = []
for pair in combinations:
    if pair["sum"] not in unique_sums:
        unique_sums.append(pair["sum"])
unique_sums.sort()

unique_sums
```

Similarly, we find all unique products

```python
unique_prods = []
for pair in combinations:
    if pair["prod"] not in unique_prods:
        unique_prods.append(pair["prod"])

unique_prods
```

Now, we find how many times does a product repeat within the combinations set

```python
prod_counts = {}
for prod in unique_prods:
    count = len([1 for p in combinations if p["prod"] == prod])
    prod_counts[prod] = count

prod_counts
```

Now, we consolidate sums factor products - for each unique sum found in the set, we determine how many factors it can be broked into and the producs of these factors

```python
sum_factors_products = {}
for s in unique_sums:
    prods = [p["prod"] for p in combinations if p["sum"] == s]
    sum_factors_products[s] = prods

sum_factors_products
```

Similarly, we can find number of factorizations for each unique product form the set and the sum of these factors

```python
product_factor_sums = {}
for prod in unique_prods:
    sums = [p["sum"] for p in combinations if p["prod"] == prod]
    product_factor_sums[prod] = sums

product_factor_sums
```

The Sum scientist has said at they knew for certain, from the very beginning, that the Prod scientist did not know the answer to the problem. This gives us an important piece of information: it means that whatever sum of the numbers was, all of the possible factors of tha sum produced an ambiguous product.

By ambiguous prouct, it is meant that here is more than one way to factorise said product.

Given this information, we can narrow down the possible sums of the numbers

```python
possible_sums = []
for s in unique_sums:
    all_products_ambiguous = True
    for prod in sum_factors_products[s]:
        if prod_counts[prod] == 1:
            all_products_ambiguous = False
            break
    if all_products_ambiguous:
        possible_sums.append(s)

possible_sums
```

Now that we have narrowed down the possible sums, we can also narrow down possible unique products - only those that can be produced from all possible factorizations of possible sums.

We can also narow that down to only those products which are only seen once through all of the possible sums. This is due to the fact that the Prod scientist could immediately guess the answer when looking at these possible sums.

```python
possible_prods = {}
for s in possible_sums:
    prods = sum_factors_products[s]
    for p in prods:
        if p in possible_prods:
            possible_prods[p] += 1
        else:
            possible_prods[p] = 1
possible_prods = [p for p in possible_prods if possible_prods[p] == 1]
possible_prods.sort()

possible_prods
```

In the end, the Sum scientist ws able to immediately guess the answer. This indicates, that out of all of the possible sums, only one of them contained exactly one possible product. Here we fund such a sum:

```python
possible_sums = [s for s in possible_sums if len([1 for p in sum_factors_products[s] if p in possible_prods]) == 1]

possible_sums
```

We now know that the sum of our numbers is 17. Now we can find the corresponding product from our list of possible products:

```python
answer_sum = possible_sums[0]
answer_sum_factors_products = sum_factors_products[answer_sum]
possible_prods = [p for p in answer_sum_factors_products if p in possible_prods]
possible_prods
```

Now that we know both the sum and the product of our numbers, we find the one combination which matches both:

```python
answer_prod = possible_prods[0]
possible_combinations = [c for c in combinations if c["sum"] == answer_sum and c["prod"] == answer_prod]
possible_combinations
```
